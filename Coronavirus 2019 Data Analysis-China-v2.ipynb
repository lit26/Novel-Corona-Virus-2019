```python
base_path = 'COVID-19/csse_covid_19_data/csse_covid_19_time_series/'
df_confirmed = cf.readData(base_path, 'time_series_19-covid-Confirmed.csv','Confirmed')
df_confirmed.head()
```

```python
df_deaths = cf.readData(base_path, 'time_series_19-covid-Deaths.csv','Deaths')
df_deaths.head()
```

```python
df_recovered = cf.readData(base_path, 'time_series_19-covid-Recovered.csv','Recovered')
df_recovered.head()
```

```python
# combine data
df = df_confirmed.append(df_deaths)
df = df.append(df_recovered)
df = df[['Date', 'Province/State', 'Country/Region', 'Confirmed','Deaths','Recovered']]
df[['Confirmed','Deaths','Recovered']] = df[['Confirmed','Deaths','Recovered']].fillna(0)
df
```

```python
#listing all the countries where the virus has spread
country_list=list(df['Country/Region'].unique())
print(country_list)
```

# Data preprocessing

```python
df_China = df[df['Country/Region']=='Mainland China']
df_China.drop('Country/Region', axis=1, inplace=True)
df_China_stat = df_China.groupby(['Date']).sum().reset_index(drop=None)
df_China_stat['Daily_confirmed_cases'] = df_China_stat.Confirmed.diff()
df_China_stat['Daily_deaths'] = df_China_stat.Deaths.diff()
df_China_stat['Daily_recovered'] = df_China_stat.Recovered.diff()
df_China_stat['Deaths_rate'] = df_China_stat.apply(lambda x: x['Deaths']*100/x['Confirmed'], axis=1)
```

# Plot graphs

```python
plt.plot('Date', 'Confirmed', data=df_China_stat, color='#D79913')
plt.xticks(rotation=50)
plt.ylabel('Number of confirmed cases in China', fontsize=15)
plt.xlabel('Date', fontsize=15)
plt.show()
```

```python
plt.plot('Date','Deaths',data=df_China_stat, color='red')
plt.plot('Date','Recovered',data=df_China_stat, color='green')
plt.xticks(rotation=50)
plt.ylabel('Number of cases', fontsize=15)
plt.xlabel('Date', fontsize=15)
plt.legend()
plt.show()
```

```python
plt.plot('Date','Deaths_rate',data=df_China_stat, color='red')
plt.xticks(rotation=50)
plt.ylabel('Death rate', fontsize=15)
plt.xlabel('Date', fontsize=15)
plt.show()
```
